```python
from scipy.ndimage import imread
import matplotlib.pyplot as plt
import scipy.misc

import os
import numpy as np

img = imread(os.path.join(directory,"RP_SPX_609.png"), flatten=True)
img[img==255] = 0
print (img.shape)

def display_image(img):
    plt.imshow(img, cmap='gray')
    plt.show()
    
display_image(img)
```

```python
def get_sub_image_y_coor(image_vec):
    '''Find the first non-zero row, and the last non-zero row - These would be y1 and y2
       Then find the first non-zero column and the last non-zero column, and this would be x1, x2'''
    y1 = 0
    for i in image_vec:
        if i == 0:
            y1 += 1
        else:
            break
    y2 = image_vec.shape[0]-1
    for i in reversed(range(image_vec.shape[0])):
        if image_vec[i] == 0:
            y2 -= 1
        else:
            break
    return y1, y2


def crop_and_save(image, file_name, output_dir):
    'Convert white pixel to black pixel'
    image[image==255] = 0
    '''Given the image, save the image with name image_name in the output_dir'''
    image_vector_vertical = image[:,int(image.shape[1]/2)]
    y1, y2 = get_sub_image_y_coor(image_vector_vertical)
    print(y1, y2)
    
    image_vector_horizontal = image[int(image.shape[1]/2), :]
    x1, x2 = get_sub_image_y_coor(image_vector_horizontal)
    print(x1,x2)
    
#     display_image(image[y1:y2, x1:x2])
    crop = image[y1:y2, x1:x2]
#     print(image.shape)
#     print(crop.shape)
    scipy.misc.imsave(os.path.join(output_dir, file_name+'.jpg'), crop)
```

```python
import re

def find_num_from_string(name):
    return re.match('\\w*?([0-9]+)\\w*', name).group(1)
```

```python
for plot in os.listdir(directory):
    my_img = imread(os.path.join(directory, plot), flatten=True)
    crop_and_save(my_img, find_num_from_string(plot), processed_plots_directory)
```
